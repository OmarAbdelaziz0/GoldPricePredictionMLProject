Finding The Correlations Between The Various Columns In The Dataset  

There are Two Types Positive Correlation and Negative Correlation
in a +ve Correlation we can assume Direct Proportionality, however, in a -ve Correlation we can assume Inverse Proportionality

```python
# Convert the 'Date' column to datetime objects
gold_data['Date'] = pd.to_datetime(gold_data['Date'])

# Extract ordinal values from datetime objects for numerical representation
gold_data['Date'] = gold_data['Date'].apply(lambda date: date.toordinal())

# Calculate the correlation matrix
correlation = gold_data.corr()
```

```python
# constructing a heatmap to understand the correlatiom
plt.figure(figsize = (6,6))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True, annot_kws={'size':10}, cmap='Reds')
```

```python
# correlation values of GLD
print(correlation['GLD'])
```

```python
# checking the distribution of the GLD Price
sns.distplot(gold_data['GLD'],color='red')
```

Splitting The Features And The Target

```python
X = gold_data.drop(['Date','GLD'],axis=1) # whenever we drop a column we need to mention that the axis is equal to 1 and the row axis=0
Y = gold_data['GLD'] # we removed the date from the dataframe and labeled GLD as our Traget in Y
```

```python
print(X) # this is the dataset without the Gold and the date. Just the Stock Prices and will be used to predict the gold price values
```

```python
print(Y)# this contains all the gold prices which as a medium remians from 100 to 140, so we have successfully splitted the features and the target
```

Splitting Into Testing Data And Training Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

Model Training: Random Forest Regressor

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# training the model
regressor.fit(X_train,Y_train)
```

Model Evaluation


```python
# prediction on Test Data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)# these are the values predicted by thye model
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

Compare the Actual Values and Predicted Values in a Plot

```python
Y_test = list(Y_test) # had to change it into a list because it gave me errors when plotting
```

```python
plt.plot(Y_test, color='red', label = 'Actual Value')
plt.plot(test_data_prediction, color='black', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
